```python
import spacy
nlp = spacy.load("en_core_web_sm")
input="this is nlp cls"
doc = nlp(input)
print([(w.text, w.pos_) for w in doc],end="\n")
```

```python
for token in doc:
  print(token)
```

```python
for sent in doc.sents:
  print(sent)
```
