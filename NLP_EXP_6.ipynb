```python
import pandas as pd
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
```

```python
text="Today is Sunday and   i'am practicing Python "
```

```python
print(text)
```

```python
nltk.download('punkt_tab')
token=word_tokenize(text)
print("Tokenization: \n")
print(token)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in token if word.lower() not in stop_words]
print("Stop-word Removal:")
print(filtered_tokens)
print("\n")
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(" Lemmatization:")
print(lemmatized_words)
print("\n")
```

```python
nltk.download('averaged_perceptron_tagger_eng') # Download the specific English tagger
pos_tags = pos_tag(token)
print(" Part-of-Speech Tagging:")
print(pos_tags)
print("\n")
```

```python
import nltk
nltk.download('maxent_ne_chunker_tab')
from nltk.chunk import ne_chunk
ner_tree = ne_chunk(pos_tags)
print("6. Named Entity Recognition:")
print(ner_tree)
```
